# Implémentation de l'algorithme de Djikstra

```python
import time

class Djikstra:
    def __init__(self, graph):
        self.graph = graph
        self.visited = {}
        self.distances = {}
        self.previous_nodes = {}
        self.temp_execution = 0

    def shortest_path(self, start, end):
        start_time = time.perf_counter()
        self.distances = {city: float('inf') for city in self.graph}
        self.distances[start] = 0
        self.visited = {}
        self.previous_nodes = {}
        self.dijkstra_recursive(start, end)

        path = self._build_path(start, end)
        end_time = time.perf_counter()
        self.temp_execution = end_time - start_time
        return self.temp_execution, path, self.distances.get(end, float('inf'))

    def dijkstra_recursive(self, current, end):
        if len(self.visited) == len(self.graph):
            return

        unvisited = {city: dist for city, dist in self.distances.items()
                     if city not in self.visited}

        if not unvisited or min(unvisited.values()) == float('inf'):
            return

        current = min(unvisited, key=unvisited.get)
        current_distance = self.distances[current]

        for neighbor, distance in self.graph[current].items():
            if neighbor not in self.visited:
                new_distance = current_distance + distance
                if new_distance < self.distances[neighbor]:
                    self.distances[neighbor] = new_distance
                    self.previous_nodes[neighbor] = current

        self.visited[current] = current_distance

        self.dijkstra_recursive(current, end)

    def _build_path(self, start, end):
        def build_path_recursive(node):
            if node == start:
                return [start]
            if node not in self.previous_nodes:
                return [node]
            return build_path_recursive(self.previous_nodes[node]) + [node]

        return build_path_recursive(end)
```

# affichage des resultats

```python
bestWay = Djikstra(graph)
time, path, distance = bestWay.shortest_path(City.Quebec, City.Buffalo)
print(f"Shortest path from Montreal to Chicago: {' -> '.join(city.name for city in path)} with distance {distance} km")
print(f"Execution time: {time:.6f} seconds")
```
