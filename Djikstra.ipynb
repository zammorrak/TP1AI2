# Algorithme de Djikstra
1. Créer un enumeration pour les villes
2. Créer une variable Graph pour représenter le graphe (nodes et arcs avec les distances)


```python
from enum import StrEnum

class City(StrEnum):
    Montreal = "a"
    Quebec = "b"
    Ottawa = "c"
    Toronto = "d"
    Buffalo = "e"
    Boston = "f"
    NewYork = "g"
    Chicago = "h"

graph = {
    City.Montreal: {
        City.Quebec: 250,
        City.Ottawa: 200 ,
        City.Boston: 435,
        City.NewYork: 595,
    },
    City.Quebec: {
        City.Montreal: 255,
        City.Boston: 650,
    },
    City.Ottawa: {
        City.Montreal: 195,
        City.Toronto: 450,
    },
    City.Toronto: {
        City.Ottawa: 445,
        City.Buffalo: 155,
        City.Chicago: 840,
    },
    City.Buffalo: {
        City.Toronto: 160,
        City.Chicago: 860,
        City.NewYork: 590,
        City.Boston: 700,
    },
    City.Boston: {
        City.Montreal: 440,
        City.Buffalo: 695,
        City.NewYork: 345,
    },
    City.NewYork: {
        City.Buffalo: 585,
        City.Boston: 350,
        City.Chicago: 1270,
    },
    City.Chicago: {
        City.Toronto: 835,
        City.Buffalo: 855,
        City.NewYork: 1275,
    }
}

print(graph)
```

```python
start_city = City.Quebec
end_city = City.Buffalo
```

# Implémentation du graphique
1. Utiliser la bibliothèque graphviz pour visualiser le graphe
2. Rajouter des attributs pour améliorer la lisibilité du graphe (couleurs, formes, etc.)
3. Afficher le graphe généré

```python
from graphviz import Digraph

def visualize_path(graph, path):
    dot = Digraph("CityGraph", format="svg")
    dot.attr(rankdir="TB")
    dot.attr("node", shape="circle", style="filled", fillcolor="lightblue", fontname="Helvetica")

    for city in City:
        dot.node(city.value, city.name)

    path_edges = set()
    for i in range(len(path) - 1):
        path_edges.add((path[i].value, path[i+1].value))

    for source, destinations in graph.items():
        for destination, distance in destinations.items():
            if (source.value, destination.value) in path_edges:
                dot.edge(source.value, destination.value,
                        label=str(distance),
                        color="red",
                        penwidth="3")
            else:
                dot.edge(source.value, destination.value,
                        label=str(distance),
                        color="lightgray")

    for city in path:
        dot.node(city.value, city.name, fillcolor="lightcoral", style="filled")

    if path:
        dot.node(path[0].value, path[0].name, fillcolor="lightgreen", style="filled")

    if path:
        dot.node(path[-1].value, path[-1].name, fillcolor="orange", style="filled")

    return dot
```

# Implémentation de la classe de djikstra
1. Créer une classe Djikstra qui prend le graphe en entrée
2. Implémenter la méthode `shortest_path(start, end)` qui retourne le chemin, la distance totale, le nombre de nœuds explorés et le temps d'exécution
3. Utiliser une approche récursive pour explorer les nœuds et mettre à jour les distances
4. Mesurer le temps d'exécution de la méthode `dijkstra_recursive` pour évaluer les performances de l'algorithme

```python
import time

class Dijkstra:
    def __init__(self, graph):
        self.graph = graph
        self.visited = {}
        self.distances = {}
        self.previous_nodes = {}
        self.temp_execution = 0

    def shortest_path(self, start, end):
        self.distances = {city: float('inf') for city in self.graph}
        self.distances[start] = 0
        self.visited = {}
        self.previous_nodes = {}
        start_time = time.perf_counter()
        self.dijkstra_recursive(start, end)
        end_time = time.perf_counter()
        self.temp_execution = end_time - start_time
        return self.temp_execution, self.get_path(start, end), self.distances.get(end, float('inf')), len(self.visited)

    def dijkstra_recursive(self, current, end):
        #creation d'un dictionnaire des nœuds non visités avec leurs distances actuelles relier aux noeuds current
        unvisited = { city: dist for city, dist in self.distances.items() if city not in self.visited }
        #si il n'y a plus de nœuds non visités ou si la distance minimale est infinie, on arrête la récursion
        if not unvisited or min(unvisited.values()) == float('inf'):
            return

        current = min(unvisited, key=unvisited.get)
        current_distance = self.distances[current]

        for neighbor, distance in self.graph[current].items():
            if neighbor not in self.visited:
                new_distance = current_distance + distance
                if new_distance < self.distances[neighbor]:
                    self.distances[neighbor] = new_distance
                    self.previous_nodes[neighbor] = current

        self.visited[current] = current_distance

        self.dijkstra_recursive(current, end)

    def get_path(self, start, end):
        path = []
        current = end

        while current is not None:
            path.append(current)
            current = self.previous_nodes.get(current)

        path.reverse()

        if path[0] != start:
            return []

        return path
```

# affichage des resultats

```python

dijkstra = Dijkstra(graph)
time, path, distance, node_explorer = dijkstra.shortest_path(start_city, end_city)
print(f"Shortest path from {start_city.name} to {end_city.name}: {' -> '.join(city.name for city in path)}")
print(f"Total distance: {distance} km")
print(f"Number of nodes explored: {node_explorer}")
print(f"Execution time: {time:.6f} seconds")

visualize_path(graph, path)
```
