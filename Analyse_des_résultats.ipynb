### Résultats obtenus

| Métrique            | A*              | Dijkstra     |
|---------------------|-----------------|--------------|
| Chemin trouvé       |         Mentionné plus bas        |

| Nœuds explorés      | 7 noeuds        | 8 noeuds     |

### Chemin solution A*
Quebec → Montreal → Ottawa → Toronto → Buffalo — **1060 km**

### Chemin solution Dijkstra
Quebec → Montreal → Ottawa → Toronto → Buffalo — **1060 km**

### Différences de performance

**A*** est un algorithme de **recherche informée** : grâce à l'heuristique
euclidienne h(n), on oriente sa recherche vers la destination et on évite
d'explorer des nœuds inutiles. L'algorithme explore donc moins de nœuds que Dijkstra
pour trouver le même chemin optimal.

**Dijkstra** est une **recherche non informée** : on explore tous les nœuds
dans l'ordre croissant de leur distance depuis la source, sans notion de
direction. Il est exhaustif mais moins efficace quand on cherche un chemin
vers une destination précise.

### Lequel utiliser, et quand ?

**Utiliser A*** quand :
- On cherche un chemin vers une **destination précise**
- On dispose de **coordonnées géographiques**
- La performance est prioritaire

**Utiliser Dijkstra** quand :
- On veut les plus courts chemins vers **tous les sommets**
- On n'a pas d'heuristique fiable disponible
- Le graphe contient des structures irrégulières difficiles à estimer

### Conclusion
Sur ce graphe de 8 villes, la différence de temps est minime vu la petite
taille du graphe. Sur un réseau de milliers de villes, l'avantage d'A*
deviendrait significatif grâce à la réduction du nombre de nœuds explorés.
